```python
def trimm_correlated(df_in, threshold):
    '''Функция, которая удяляет из датасета корелированные признаки. threshold - это 
    предельное значение кореляции. '''
    df_corr = df_in.corr()
    df_not_correlated = ~(df_corr.mask(np.tril(np.ones([len(df_corr)]*2, dtype=bool))) > threshold).any()
    un_corr_idx = df_not_correlated.loc[df_not_correlated[df_not_correlated.index] == True].index
    df_out = df_in[un_corr_idx]
    return df_out
```

```python
import scipy.stats as scs
```

```python
def print_statistics(array):
    '''Вывод основных статистических показателей'''
    sta = scs.describe(array)
    print('%14s %15s' % ('Статистика', 'Значение'))
    print(30 * '-')
    print('%14s %15.5f' % ('size', sta[0]))
    print('%14s %15.5f' % ('min', sta[1][0]))
    print('%14s %15.5f' % ('max', sta[1][1]))
    print('%14s %15.5f' % ('mean', sta[2]))
    print('%14s %15.5f' % ('std', np.sqrt(sta[3])))
    print('%14s %15.5f' % ('skew', sta[4]))
    print('%14s %15.5f' % ('kurtosis', sta[5]))
    
```

```python
# функция теста на нормальность
def normality_tests(array):
    '''Проверка нормального распределения набора данных'''
    print('Коэффициент ассиметрии набора данных %14.3f'% scs.skew(array))
    print('P-значение коэффициента ассиметрии %14.3f'% scs.skewtest(array)[1])
    print('Коэффициент эксцесса набора данных %14.3f'% scs.kurtosis(array))
    print('P-значение коэффицента эксцесса %14.3f'% scs.kurtosistest(array)[1])
    print('P-значение критерия нормальности %14.3f'% scs.normaltest(array)[1])
```

```python
def port_ret(weights):
    '''ожидаемая годовая доходность'''
    return np.sum(rets.mean() * weights) * 252

def port_vol(weights):
    '''дисперсия'''
    return np.sqrt(np.dot(weights.T, np.dot(rets.cov()*252, weights)))
```

```python
prets = []
pvols = []
noa = len(rets.columns)
# Моделирование весов портфеля по методу Монте-Карло
for p in range(10000):    
    weights = np.random.random(noa)
    weights /= np.sum(weights)
# Сохранение статистических показателей в списках
    prets.append(port_ret(weights))  # доходность
    pvols.append(port_vol(weights))  # дисперсия (волатильность)
# сохранение в массив numpy
prets = np.array(prets)
pvols = np.array(pvols)
```

```python
plt.figure(figsize=(12,6))
plt.scatter(pvols, prets, c=prets / pvols, marker ='o', cmap='coolwarm')

plt.xlabel('Ожидаемая волотильность')
plt.ylabel('Ожидаемая доходность')
plt.colorbar(label = 'Коэффицент Шарпа')
```

## Максимизация доходности

```python
# Оптимизация портфеля
import scipy.optimize as sco

# Минимизируемая функция
def min_func_sharpe(weights): 
    return -port_ret(weights) / port_vol(weights)

# ограничение заданное равенством
cons = ({'type' : 'eq', 'fun' : lambda x: np.sum(x)-1})

# предельные значения параметров
bnds = tuple((0, 1) for x in range(noa))

# вектор равных весов
eweights = np.array(noa *[1. /noa,])

min_func_sharpe(eweights)

opts = sco.minimize(min_func_sharpe, eweights, method='SLSQP', bounds=bnds, constraints=cons)

# оптимальные веса финансовых инструментов в портфеле
max_value_portf = opts['x'].round(3)

# результирующая доходность портфеля
max_roi = port_ret(max_value_portf).round(3)

# результирующая волатильность
res_volat = port_vol(max_value_portf)

# Максимальный коэффициент Шарпа
max_sharp = max_roi / res_volat

print('*'*16, 'Результаты максимизации доходности портфеля', '*'*16)
print()
print()
print('Веса с максимальной доходностью:', max_value_portf)
print('Результирующая макс доходность:', max_roi)
print('Результирующая волатильность:', res_volat )
print('Максимальный коэффицент Шарпа:', max_sharp )


```
